## Vector Matrix Addition

```python
### 2D Matrix addition function

def matrix_add_2d(a, b):
  rows, cols = len(a), len(a[0])
  new_matrix = []
  for i in range(rows):
    new_matrix.append([])
    for j in range(cols):
      new_matrix[i].append(a[i][j] + b[i][j])
  return new_matrix
```

```python
%%timeit

a = [[ 4 , 2 , 6 ] ,[ 1 , 9 , 0 ]]

b = [[ 1 , 0 , 0 ] ,[ 4 , 0 , 3 ]]

%timeit -n 10 matrix_add_2d(a, b)
```

```python
%%timeit

## testing with larger a/b's

a= [[1,2,3,4,5,6,7,8,9],[1,2,3,4,5,6,7,8,9]]

b = [[9,8,7,6,5,4,3,2,1],[9,8,7,6,5,4,3,2,1]]

matrix_add_2d(a,b)
```

```python
import numpy as np

vector_a = np.array(a)
vector_b = np.array(b)
```

```python
%%timeit

np.add(vector_a, vector_b)
```
